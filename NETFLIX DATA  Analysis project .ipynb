```python
# My suggested sulution is to show top 10 presented genre 

movie_by_year_genre=movie_by_year.groupby(['Genre']).agg({"Title":"nunique"}).reset_index().sort_values(by=['Title'],ascending=False)[:10]
plt.figure(figsize=(15,6))
sns.barplot(x = "Genre",y = 'Title', data = movie_by_year_genre)
plt.xticks(rotation = 60)
plt.title('Top 10 Genres')
plt.show()
```
